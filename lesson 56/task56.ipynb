```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as clr
```

```python
image = cv2.imread('lyusya.jpg')
plt.imshow(image)
cv2.waitKey(0)
plt.axis('off')
plt.show
```

```python
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(image_RGB)
plt.show()
```

# Декомпозиція по RGB

```python
RED = image_RGB.copy()
RED[:, :, (1, 2)] = 0
plt.axis('off')
plt.imshow(RED)
plt.show()
```

```python
GREEN = image_RGB.copy()
GREEN[:, :, (0, 2)] = 0
plt.axis('off')
plt.imshow(GREEN)
plt.show()
```

```python
BLUE = image_RGB.copy()
BLUE[:, :, (1, 0)] = 0
plt.axis('off')
plt.imshow(BLUE)
plt.show()
```

# Декомпозиція по HSV

```python
image_HSV = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

```python
H = image_HSV.copy()
H[:, :, (1, 2)] = 255
H = cv2.cvtColor(H, cv2.COLOR_HSV2RGB)
plt.axis('off')
plt.imshow(H)
plt.show()
```

```python
S = image_HSV.copy()
S[:, :, 0] = 0
S[:, :, 2] = 255
S = cv2.cvtColor(S, cv2.COLOR_HSV2RGB)
plt.axis('off')
plt.imshow(S)
plt.show()
```

```python
V = image_HSV.copy()
V[:, :, 1] = 0
V[:, :, 0] = 179
V = cv2.cvtColor(V, cv2.COLOR_HSV2RGB)
plt.axis('off')
plt.imshow(V)
plt.show()
```

# Декомпозиція по CIE Lab

```python
image_lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
```

```python
L = image_lab.copy()
L[:, :, (1, 2)] = 0
L = cv2.cvtColor(L, cv2.COLOR_LAB2BGR)
L = cv2.cvtColor(L, cv2.COLOR_RGB2GRAY)
plt.axis('off')
plt.imshow(L)
plt.show()
```

```python
a = image_lab.copy()
a[:, :, 0] = 127
a[:, :, 2] = 0
a = cv2.cvtColor(a, cv2.COLOR_LAB2BGR)
a = cv2.cvtColor(a, cv2.COLOR_RGB2GRAY)
cmap_a = clr.LinearSegmentedColormap.from_list('custom_blue', ['Red', 'Gray', 'Green'], N = 255)
plt.axis('off')
plt.imshow(a, cmap_a)
plt.show()
```

```python
b = image_lab.copy()
b[:, :, 0] = 127
b[:, :, 1] = 0
b = cv2.cvtColor(b, cv2.COLOR_LAB2BGR)
b = cv2.cvtColor(b, cv2.COLOR_RGB2GRAY)
cmap_b = clr.LinearSegmentedColormap.from_list('custom_blue', ['Black', 'Yellow', 'Blue'], N = 255)
plt.axis('off')
plt.imshow(b, cmap_b)
plt.show()
```
