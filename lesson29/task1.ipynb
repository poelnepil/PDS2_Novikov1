**Протягом 50 років вимірювався рівень води навесні під час повені відносно умовного нуля. Результати вимірювання наведено у вигляді інтервального статистичного розподілу:**

| xi  | 0-24  | 24-48 | 48-72 | 72-96 | 96-120| 120-144 | 144-168 | 168-192 | 192-216|
|:--- | :---: | :---: | :---: | :---: | :---: | :---:   | :---:   | :---:   | :---:  |
|  ni |   1   |   2   |   4   |   6   |   12  |   16    |   6     |    2    |    1   |

```python
from decimal import *

getcontext().prec = 7

a = 0.01
x = [[0, 24], [24, 48], [48, 72], [72, 96], [96, 120], [120, 144], [144, 168], [168, 192], [192, 216]]
ni = [1, 2, 4, 6, 12, 16, 6, 2, 1]
xi = []
z = []
Fzi_1 = [-0.5, -0.4913, -0.4599, -0.3686, -0.1879, 0.0517, 0.2764, 0.4177, 0.4783]
Fzi = [-0.4913, -0.4599, -0.3686, -0.1879, 0.0517, 0.2764, 0.4177, 0.4783, 0.5]
pi = []
ni_ = []
n = sum(ni)
Kemp = 0

suma = 0
suma1 = 0



for i in x:
    xi.append((i[0]+i[1])//2)

for i in range(len(xi)):
    suma += xi[i]*ni[i]

for i in range(len(xi)):
    suma1 += (xi[i]**2)*ni[i]

x_ = (Decimal(1)/Decimal(n)) * suma

D_ = (Decimal(1)/Decimal(n)) * suma1 - x_**2

b_ = D_.sqrt()

for i in x:
    zi = ((i[1])-x_)/b_
    z.append(zi.quantize(Decimal('1.00')))
z = z[:8]

for i in range(len(Fzi_1)):
    pi.append(Fzi[i]-Fzi_1[i])

for i in range(len(pi)):
    ni_.append(pi[i]*50)

for i in range(len(ni_)):
    if ni_[i] == 0:
        ni_[i] = 1
    Kemp += ((ni[i]-ni_[i])**2)/ni_[i]

m = len(x)
S = 2
r = m - S - 1

Kkr = 16.8

if Kemp <= Kkr:
    print(True)
else:
    print(False)

print()

print(x)
print(xi)
print(ni)
print(n)
print(x_)
print(D_)
print(b_)
print(z)
print(Fzi)
print(Fzi_1)
print(pi)
print(ni_)
print(Kemp)
print(m)
print(S)
print(r)
print(Kkr)
```
