```python
import pandas as pd
import numpy as np
from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_blobs
from sklearn.mixture import GaussianMixture
```

```python
weight1 = 13
std1 = 0.5
sample1 = np.random.normal(weight1, std1, 8000)

weight2 = 12
std2 = 0.7
sample2 = np.random.normal(weight2, std2, 5000)
```

```python
data = np.concatenate((sample1, sample2))
data
```

```python
x1, y1 = make_blobs(n_samples = 8000, centers=1, cluster_std=0.5, center_box=(13,13))
x2, y2 = make_blobs(n_samples = 5000, centers=1, cluster_std=0.7, center_box=(12,12))
plt.scatter(x1[:, 0], x1[:, 1], color='red', label='sample 1')
plt.scatter(x2[:, 0], x2[:, 1], color='blue', label='sample 2')
plt.legend()
plt.show()
```

```python
gm = GaussianMixture(n_components=2).fit(data.reshape(-1, 1))
gm.get_params()
```

```python
pred = gm.predict(data.reshape(-1, 1))
```

```python
first_factory = list(pred).count(0)/13000
second_factory = list(pred).count(1)/13000
```

```python
print(f'ймовірність, що деталь з першого заводу: {first_factory}')
print(f'ймовірність, що деталь з другого заводу: {second_factory}')
```
