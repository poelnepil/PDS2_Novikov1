**Є можливість вибрати спосіб виробництва і реалізації двох наборів товарів широкого вжитку. За даними відділу маркетингу, яким були проведені дослідження ринку, можливий прибток від виробництва і реалізації Х і Y наведено в таблицях (Х, Y – прибуток у грошових одиницях).**
**Потрібно оцінити ступінь ризику і запропонувати рішення щодо випуску і реалізації одного з наборів товару.**

|  X   | 1000  | 1500  | 2000   |
|:---  | :---: | :---: | :---:  |
|  P   |  0,5  |  0,3  |  0,2   |


|  Y   | 1000  | 1500  | 1750   |
|:---  | :---: | :---: | :---:  |
|  P   |  0,4  |  0,4  |  0,2   |

***Розв'язок:***

$
M[X]:(0.5 \cdot 1000) + (0.3 \cdot 1500) + (0.2 \cdot 2000) = 1350$ - **очікуання**

$
M[X^2]:(1000^2 \cdot 0.5) + (1500^2 \cdot 0.3) + (2000^2 \cdot 0.2) = 1975000 $

$
D[X] = M[X^2] - M[X]^2 = 1975000 - 1350^2 = 152500$ - **дисперсія**

$
\delta[X] = \sqrt{D(X)} = \sqrt{152500} = 390.5 $ - **похибка**

--------------------------------------------------------------------------------------

$
M[Y]:(0.4 \cdot 1000) + (0.4 \cdot 1500) + (0.2 \cdot 1750) = 1350$ - **очікуання**

$
M[Y^2]:(1000^2 \cdot 0.4) + (1500^2 \cdot 0.4) + (1750^2 \cdot 0.2) = 1912500 $

$
D[Y] = M[Y^2] - M[Y]^2 = 1912500 - 1350^2 = 90000$ - **дисперсія**

$
\delta[Y] = \sqrt{D(Y)} = \sqrt{90000} = 300 $ - **похибка**
