```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
im_src = cv2.imread('chess_2.png')

plt.figure(figsize=(12,8))
plt.imshow(im_src)
plt.show()
```

```python
A = [216,68]
B = [472, 115]
C = [381, 347]
D = [222, 312]
pts_src = np.array([A, B, C, D])
```

```python
im_dst = cv2.imread('chess_1.png')

plt.figure(figsize=(12,8))
plt.imshow(im_dst)
plt.show()
```

```python
A = [447,72]
B = [444, 230]
C = [214, 219]
D = [210, 79]
pts_dst = np.array([D,A,B,C])
```

```python
h, status = cv2.findHomography(pts_src, pts_dst,)
```

```python
im_out = cv2.warpPerspective(im_src, h, (im_dst.shape[1], im_dst.shape[0]))

plt.figure(figsize=(12,8))
plt.imshow(im_out)
plt.show()
```
