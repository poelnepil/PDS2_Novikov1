```python
from PIL import Image, ImageEnhance
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.colors as clr
```

# Реалізуйте функцію для зчитування зображення з файлу та його відображення за допомогою бібліотеки OpenCV.



```python
def read(img):
    img = cv2.imread(img)
    img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img_RGB)
    plt.axis('off')
    plt.show()
    return img_RGB
```

```python
image = read("lyusya2.jpg")
```

# Реалізуйте фільтр Собеля для зображення та відобразіть його результат. 

```python
image0 = cv2.imread('lyusya2.jpg')

gray = cv2.cvtColor(image0, cv2.COLOR_RGB2GRAY)

ddept=cv2.CV_16S

x = cv2.Sobel(gray, ddept, 1,0, ksize=3, scale=1)
y = cv2.Sobel(gray, ddept, 0,1, ksize=3, scale=1)

absx = cv2.convertScaleAbs(x)
absy = cv2.convertScaleAbs(y)

edge = cv2.addWeighted(absx, 0.5, absy, 0.5, 0)

plt.imshow(edge, cmap='gray')
plt.axis('off')
plt.show()
```

# Реалізуйте медіанний фільтр для зображення та відобразіть його результат.


```python
image1 = cv2.imread('lyusya2.jpg')

dst = cv2.medianBlur(image1,5)

dst_RGB = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)

plt.axis('off')
plt.imshow(dst_RGB)
plt.show()
```

# Порівняння фільтрів

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(121)
plt.imshow(edge, cmap='gray')
plt.title('Sobel')
plt.axis('off')

ax = fig.add_subplot(122)
plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))
plt.title('Median')
plt.axis('off')

plt.show()
```

**Медіанний фільтр та фільтр Собеля - це два різні методи обробки зображень, які мають різні застосування. Медіанний фільтр зменшує шум, тоді як фільтр Собеля виявляє границі на зображенні.**
